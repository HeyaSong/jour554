# 1. Electricity Supply-Demand Balance-Annual Data

####### Link:
National Bureau of Statistics of China:https://data.stats.gov.cn/english/easyquery.htm?cn=C01
BP Statistics (just for double check):https://www.bp.com/en/global/corporate/energy-economics/statistical-review-of-world-energy.html
####### Description:Utilizing China's national data, I would like to look into the trend of electricity supply-demand balance of China in past 20 years, laying a good foundation for further exploration of 2021 situation

### read data

```python
df = pd.read_csv("./annual.csv")
df.head()
```

### transpose rows and columns

```python
df_t = df.T
df_t.head(22)
```

### 🌟reorganize table

```python
# make the first row as index
df_new = pd.DataFrame(df_t.values[1:], columns=df_t.values[0])
```

```python
# drop empty row
df_new = df_new.drop([0])
```

```python
# extract the data that I need
df_new = df_new.loc[
    :,
    [
        "Indicators",
        "Outputof Electricity(100 million kw.h)",
        "Total Electricity Consumption(100 million kw.h)",
    ],
]
```

```python
# rename columns
df_new.columns = ["Year", "Production", "Consumption"]
```

```python
# change data types
df_new["Year"] = pd.to_datetime(df_new["Year"].astype("int").astype("str")).astype(
    "datetime64[Y]"
)
df_new["Production"] = df_new["Production"].astype("float")
df_new["Consumption"] = df_new["Consumption"].astype("float")
```

```python
# reorder years
df_new = pd.DataFrame(df_new.sort_values(by="Year", ascending=True))
df_new
```

```python
# Finally good to go！！ 😭
```

### draw chart

```python
# using Altair draw Layered Area Chart for "Outputof Electricity(100 million kw.h)" & "Total Electricity Consumption(100 million kw.h)"
# https://altair-viz.github.io/gallery/layered_area_chart.html

## First Drawing
base_figure = alt.Chart(df_new).encode(x="Year")
alt.layer(
    base_figure.mark_area(opacity=0.3, color="blue").encode(y="Production:Q"),
    base_figure.mark_area(opacity=0.3, color="red").encode(y="Consumption:Q"),
)
## blue+red=purlple...the result is too similar to be observed
```

```python
# Calculate the Discrepancy so that I can draw the difference
df_new["P-C"] = df_new["Production"] - df_new["Consumption"]
```

```python
## Second Drawing
base_figure = alt.Chart(df_new).encode(x="Year")
alt.layer(base_figure.mark_area(opacity=0.3, color="green").encode(y="P-C:Q"))
```

# 2. Electricity Supply-Demand Balance-Monthly Data

### clean data

```python
df_2 = pd.read_csv("./monthly.csv")
```

```python
df_2.columns = df_2.columns.str.strip().str.lower()
```

```python
df_2["electricity generation twh"] = (
    df_2["electricity generation twh"]
    .str.replace("-- Hydro", "Hydro", regex=False)
    .str.replace("-- Thermal", "Thermal", regex=False)
    .str.replace("-- Nuclear", "Nuclear", regex=False)
)
```

```python
df_2
```

```python
df_2.keys()
```

```python
# rename columns
df_2.columns = ["ge_m"] + [_ for _ in df_2.columns[1:]]
df_2.head(22)
```

```python
# change data types
df_2["value"] = df_2["value"].astype("float")
```

### 🌟 extract the data needed (while expression)

```python
import datetime

# 1. Build DataFrame for "generation - consumption"
df_2_gc_dict = {"month": [], "value": []}
index = 0
while index < len(df_2):
    df_2_gc_dict["month"].append(datetime.datetime.strptime(df_2["month"][index], "%B"))
    # new value = original generation - original consumption
    df_2_gc_dict["value"].append(df_2["value"][index] - df_2["value"][index + 4])
    index += 16
df_2_gc = pd.DataFrame.from_dict(df_2_gc_dict)
df_2_gc
```

```python
# 2. Build DataFrame for Different Types of Energy
df_2_type_dict = {"month": [], "type": [], "value": []}
index = 1
while index < len(df_2):
    df_2_type_dict["month"].append(
        datetime.datetime.strptime(df_2["month"][index], "%B")
    )
    df_2_type_dict["type"].append("Hydro")
    df_2_type_dict["value"].append(df_2["value"][index])
    df_2_type_dict["month"].append(
        datetime.datetime.strptime(df_2["month"][index], "%B")
    )
    df_2_type_dict["type"].append("Thermal")
    df_2_type_dict["value"].append(df_2["value"][index + 1])
    df_2_type_dict["month"].append(
        datetime.datetime.strptime(df_2["month"][index], "%B")
    )
    df_2_type_dict["type"].append("Nuclear")
    df_2_type_dict["value"].append(df_2["value"][index + 2])
    index += 16
df_2_type = pd.DataFrame.from_dict(df_2_type_dict)
df_2_type
```

```python
# 1. Draw generation - consumption
base_figure = alt.Chart(df_2_gc).encode(x="month:T")
alt.layer(base_figure.mark_line(opacity=0.3, color="blue").encode(y="value:Q"))
```

```python
# 2. Draw various types of source
alt.Chart(df_2_type).mark_area().encode(x="month:T", y=alt.Y("value:Q"), color="type:N")
```

```python
## 2. Draw various types of source
alt.Chart(df_2_type).mark_area().encode(x="month", y=alt.Y("value:Q"), color="type:N")
```

# 3.Power Resoucres

####### Link:https://en.wikipedia.org/wiki/Electricity_sector_in_China#cite_note-:5-26
####### Description:with "Electricity production (GWh) in China by source, 2008-2020" from Wikipedia, I will calculate the percentage of each category (fossil, nuclear and renewable) among the total. Then I would like to draw a normalized stacked area chart to see the trend of proportion of electricity generated by different categories

### web script

```python
url = "https://en.wikipedia.org/wiki/Electricity_sector_in_China#cite_note-:5-26"
```

```python
page = pd.read_html(url, attrs={"class": "wikitable sortable"})
```

```python
df_3 = pd.DataFrame(page[0])
```

```python
df_3.columns
```

### 🌟 deal with multiIndex issue

```python
# to progress to the next step, I have to convert MultiIndex into a Single one
lst = df_3.values
df_3 = pd.DataFrame(lst, columns=[_[0] for _ in df_3.keys()])
df_3.head()
```

### reorganize the table

```python
df_3 = df_3.drop(columns=["% renewable"])
df_3 = df_3.drop([12])
```

```python
df_3.columns = (
    ["Year_source"]
    + ["Total"]
    + ["Fossil_1"]
    + ["Fossil_2"]
    + ["Fossil_3"]
    + ["Nuclear"]
    + ["Hydro"]
    + [_ for _ in df_3.columns[7:]]
)
df_3
```

```python
df_3 = df_3.apply(pd.to_numeric)
```

```python
df_3.info()
```

```python
df_3["Year_source"] = pd.to_datetime(df_3["Year_source"].astype(str)).astype(
    "datetime64[Y]"
)
```

```python
df_3["Total fossil"] = df_3["Fossil_1"] + df_3["Fossil_2"] + df_3["Fossil_3"]
```

```python
df_3new = df_3.loc[
    :, ["Year_source", "Total fossil", "Nuclear", "Hydro"],
]
df_3new
```

### Draw a chart

```python
base_3figure = alt.Chart(df_3new).encode(x="Year_source")
alt.layer(
    base_3figure.mark_area(opacity=0.3, color="orange").encode(y="Total fossil:Q"),
    base_3figure.mark_area(opacity=0.3, color="black").encode(y="Nuclear:Q"),
    base_3figure.mark_area(opacity=0.3, color="blue").encode(y="Hydro:Q"),
)
```
