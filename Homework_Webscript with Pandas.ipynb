```python
url = 'https://www.realclearpolitics.com/epolls/other/kamala_harris_favorableunfavorable-6690.html#polls'
```

## Read table

```python
df = pd.read_html(url)[1]
```

## Clean up repeated name

```python
df['Poll']=df['Poll'].str.replace("Economist/YouGovYouGov",'Economist').replace('Politico/Morning ConsultPolitico','Politico').replace('FOX NewsFOX News','FOX News').replace('QuinnipiacQuinnipiac','Quinnipiac').replace('Reuters/IpsosReuters','Reuters')
```
